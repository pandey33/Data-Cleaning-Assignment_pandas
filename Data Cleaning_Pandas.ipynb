```python
#Solution 1:
    
df['FlightNumber'] = df['FlightNumber'].replace(np.nan,0)
df.astype({'FlightNumber': 'int64'}).dtypes
df1 = df['FlightNumber'] 

for i in df1:
    if i != 0:
        a = i
    if i == 0.0:
        i = a + 10.0

    print (i)
```

```python
#new data frame  df2 with split value columns
df2 = df['From_To'].str.split("_", n = 1, expand = True)
df2
```

```python
#solution 2

# making separate From column from new data frame
df['From'] = df2[0]

# making separate To column from new data frame
df['To'] = df2[1]

# Dropping old Name columns 
df.drop(columns =['From_To'], inplace = True) 

df
```

```python
#Solution 3 & 4

df['From'] = df['From'].str.title()
df['To'] = df['To'].str.title()


df
```

```python
#Solution 5

#Finding the maximum no. of list value for creating the no. of columns
x =[]
for i in df['RecentDelays']:
    x.append(len(i))
    y = max(x)
print('max no. of values in list of column RecentDelays: {}' .format(y))

print ('-' *40)
print (' ' *40)


#Creating new columns based on maximum no. of value found in list
p=[]
for i in range(y):
    i = str(i)
    p.append('delay'+i)
print ('Name of new columns that needs to be added in dataframe are: {}' .format(p))

print ('-' *40)
print (' ' *40)


#Seggrating the list value in respective newly created columns
q = pd.DataFrame(df['RecentDelays'].to_list(), columns=p)
print ('New column values as: \n {}' .format(q))

print ('-' *40)
print (' ' *40)

#concating the q dataframe with df on the same indexes using keyword axis =1
df = pd.concat([df,q], axis=1)
df

print ('-' *40)
print (' ' *40)


# Dropping old Name columns 
df.drop(columns =['RecentDelays'], inplace = True) 

print ('Final Output: \n ')

df

```
